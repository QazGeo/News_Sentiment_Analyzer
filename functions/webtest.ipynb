```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
url = "https://bbc.com/sports"
response = requests.get(url)
content = response.text
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
header =soup.find_all(['h1', 'h2', 'h3'])
for head in header:
    h1text = head.get_text()
    print(h1text)
```

```python
article = soup.find_all(['p'])
data = []
num = 0
for text in article:
    content = text.get_text(strip=True)
    if len(content.split()) > 5:
        data.append(content)
    num += 1
    print(num, ". " + content)
```

```python
print(len(data))
print(data)
```

```python
df = pd.DataFrame(data, columns=['content'])

csv_file = 'extract.csv'
df.to_csv(csv_file, index=False, encoding='utf-8-sig')
```
